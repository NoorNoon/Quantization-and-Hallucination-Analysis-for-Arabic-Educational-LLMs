## Create prompts.py

### Subtask:
Define functions for zero-shot, few-shot with QA, and few-shot with CoT prompting strategies.

#### Instructions

**Reasoning**:
I will create the `prompts.py` file and implement the `get_prompt_template` function along with example data for few-shot prompting, as outlined in the subtask instructions. This ensures all required prompting strategies are available for evaluation.



````python
%%writefile prompts.py
import textwrap
from typing import Dict, List, Optional

# Main prompt types for primary evaluation
MAIN_PROMPT_TYPES = ["simple", "long", "cot", "instruction"]

# Bonus prompt types for optional side experiments
BONUS_PROMPT_TYPES = ["few_shot_qa", "few_shot_cot"]

# Combined list of all prompt types
PROMPT_TYPES = MAIN_PROMPT_TYPES + BONUS_PROMPT_TYPES


# Fixed examples for few-shot QA prompting strategy
FEW_SHOT_QA_EXAMPLES = [
    {"question": "ما هي عاصمة السعودية؟", "answer": "الرياض."},
    {"question": "ما هو أكبر كوكب في المجموعة الشمسية؟", "answer": "المشتري."},
    {"question": "ما هو ناتج 5 + 3؟", "answer": "8."},
]

# Fixed examples for few-shot CoT (Chain-of-Thought) prompting strategy
FEW_SHOT_COT_EXAMPLES = [
    {
        "question": "إذا كان سعر كتاب 10 ريالات وسعر قلم 2 ريال، ولقد اشتريت كتابين وثلاثة أقلام، فكم مجموع ما دفعته؟",
        "thought": "تكلفة الكتب = 2×10=20. تكلفة الأقلام = 3×2=6. المجموع = 20+6=26.",
        "answer": "26 ريال."
    },
    {
        "question": "ما هي مراحل دورة الماء في الطبيعة؟",
        "thought": "تبخر الماء ثم تكثفه لتكوين السحب ثم الهطول ثم تجمعه.",
        "answer": "تبخر، تكثف، هطول، وتجمع."
    },
]


def build_prompts(
    question: str,
    few_shot_qa: Optional[List[Dict]] = None,
    few_shot_cot: Optional[List[Dict]] = None,
) -> Dict[str, str]:
    """
    Generates a dictionary of prompts for a given question using different styles.
    The prompt style changes, but the question remains constant.

    Designed for research evaluation, ensuring:
    - CoT prompts do not show internal thoughts.
    - Few-shot prompts do not repeat examples or introduce additional questions.
    - Model provides a direct answer (with 'الإجابة:' / 'الإجابة النهائية:').
    """
    q = question.strip()

    # Main prompts for the primary evaluation
    prompts = {
        "simple": textwrap.dedent(f"""
        أجب عن السؤال التالي بإيجاز ودقة.
        السؤال: {q}
        الإجابة:
        """).strip(),

        "long": textwrap.dedent(f"""
        أجب عن السؤال التالي بشرح مبسط لكن مفصل، واذكر مثالاً واحداً إن أمكن.
        لا تذكر مصادر ولا روابط ولا نصائح بحث.
        السؤال: {q}
        الإجابة:
        """).strip(),

        "cot": textwrap.dedent(f"""
        أجب عن السؤال التالي.
        فكّر خطوة بخطوة داخلياً (لا تكتب خطواتك).
        لا تذكر مصادر، ولا روابط، ولا تقترح البحث على الإنترنت.

        السؤال: {q}
        الإجابة النهائية:
        """).strip(),

        "instruction": textwrap.dedent(f"""
        أنت معلّم محترف. أجب بالعربية الفصحى بأسلوب تعليمي مبسط.
        لا تذكر مصادر أو روابط أو نصائح بحث.
        قدّم الإجابة مباشرة.
        السؤال: {q}
        الإجابة:
        """).strip(),
    }

    # Bonus few-shot prompts
    if few_shot_qa is None:
        few_shot_qa = FEW_SHOT_QA_EXAMPLES
    if few_shot_cot is None:
        few_shot_cot = FEW_SHOT_COT_EXAMPLES

    ex_qa = "\n".join([f"السؤال: {ex['question']}\nالإجابة: {ex['answer']}" for ex in few_shot_qa])

    # Few-shot QA prompt: answers only the last question, no example repetition
    prompts["few_shot_qa"] = textwrap.dedent(f"""
    لديك أمثلة (سؤال/إجابة). مهم:
    - أجب فقط عن السؤال الأخير (سؤال المستخدم).
    - لا تكرر الأمثلة.
    - لا تكتب أي أسئلة إضافية.
    - أعد الإجابة فقط بسطر واحد.

    {ex_qa}

    سؤال المستخدم: {q}
    الإجابة:
    """).strip()

    ex_cot = "\n".join([
        f"السؤال: {ex['question']}\nتفكير مختصر: {ex['thought']}\nالإجابة: {ex['answer']}"
        for ex in few_shot_cot
    ])

    # Few-shot CoT prompt: thinks internally, no written thoughts
    prompts["few_shot_cot"] = textwrap.dedent(f"""
    لديك أمثلة تتضمن (تفكير مختصر/إجابة). مهم:
    - أجب فقط عن سؤال المستخدم الأخير.
    - لا تكرر الأمثلة.
    - فكّر داخلياً (لا تكتب التفكير).
    - لا تذكر مصادر أو روابط أو نصائح بحث.
    - اكتب الإجابة النهائية فقط بسطر واحد أو سطرين.

    {ex_cot}

    سؤال المستخدم: {q}
    الإجابة النهائية:
    """).strip()

    return prompts


print("prompts.py created successfully.")
````

### Subtask:
Implement a Python script (`run_eval.py`) to evaluate language models with different precisions (FP16, 4-bit quantized) and prompting strategies (zero-shot, few-shot QA, few-shot CoT) on Arabic educational questions.

    -   For `fp16` precision, load with `torch_dtype=torch.float16`.

7.  **Dependencies**: Ensure necessary imports like `torch`, `transformers`, `argparse`, `json`, `os`, and `prompts` are included.

**Reasoning**:
I will create the `run_eval.py` script based on the detailed instructions provided in the subtask. This script will handle argument parsing, model loading for different precisions, loading questions, iterating through them to generate model responses using various prompt types, and finally saving the results to a JSONL file.



````python
%%writefile run_eval.py
# run_eval.py
# Project: Prompt sensitivity evaluation (Arabic) — with model output post-processing

import argparse
import json
import time
import pandas as pd
import torch
import math
from typing import Tuple

from transformers import AutoTokenizer, AutoModelForCausalLM
from prompts import build_prompts, PROMPT_TYPES


def read_questions_jsonl(path: str):
    """Reads questions from a JSONL file."""
    questions = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            questions.append(json.loads(line))
    return questions


def pick_device(device: str) -> str:
    """Picks the appropriate device (cuda or cpu)."""
    if device == "auto":
        return "cuda" if torch.cuda.is_available() else "cpu"
    return device


def clean_output(text: str) -> str:
    """
    Strict cleaning of model outputs:
    - Truncates at dialogue/role markers (Human/Assistant/User/System).
    - Removes "thought/steps" sections if present.
    - Limits text to a maximum of two lines.
    - Returns clean text suitable for evaluation and display.
    """
    if not text:
        return ""

    t = text.strip()

    # 1) Dialogue/role markers — truncate at first occurrence
    cut_markers = [
        "Human:", "Assistant:", "USER:", "SYSTEM:",
        "المستخدم:", "المساعد:", "النظام:",
        "###", "---", "Human :", "Assistant :"
    ]
    for m in cut_markers:
        if m in t and not t.startswith(m):
            t = t.split(m, 1)[0].strip()

    # 2) Question/example markers (prevents repeating few-shot examples)
    example_markers = [
        "السؤال:", "سؤال المستخدم:", "سؤال:", "أمثلة:", "مثال:"
    ]
    for m in example_markers:
        if m in t and not t.startswith(m):
            # Truncate at the next question or repeated example section
            t = t.split(m, 1)[0].strip()

    # 3) Remove "thought/steps" sections
    step_markers = [
        "الخطوات", "خطوات", "فكرة الخطوات", "تفكير", "التفكير", "الخطوات الداخلية", "خطوة", "خطوات:"
    ]
    for m in step_markers:
        if m in t:
            t = t.split(m, 1)[0].strip()

    # 4) Remove common annoying prefixes
    prefixes_to_strip = [
        "إجابة:", "Answer:", "الإجابة:", "Result:", "النتيجة:"
    ]
    for p in prefixes_to_strip:
        if t.startswith(p):
            t = t[len(p):].strip()

    # 5) Reduce multiple newlines and repetitive lines
    t = t.replace("\r\n", "\n").replace("\r", "\n")
    while "\n\n" in t:
        t = t.replace("\n\n", "\n")
    t = t.strip()

    # 6) Keep a maximum of two lines (suitable for poster display)
    lines = [ln.strip() for ln in t.splitlines() if ln.strip()]
    if len(lines) == 0:
        return ""
    t = "\n".join(lines[:2]).strip()

    # 7) Finally: if the text is too long, truncate at 300 characters (protection)
    if len(t) > 300:
        t = t[:297].rstrip() + "..."

    return t


@torch.inference_mode()
def generate_once(model, tokenizer, prompt: str, device: str, max_new_tokens: int) -> Tuple[str, float, float]:
    """
    Generates text from the model.
    Returns: (raw_text, latency_ms, peak_gpu_mem_mb_or_None)
    """
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    # Reset GPU memory statistics
    if device == "cuda":
        try:
            torch.cuda.reset_peak_memory_stats()
            torch.cuda.synchronize()
        except Exception:
            pass

    start = time.perf_counter()

    try:
        output = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=0.0,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )
    except RuntimeError as e:
        # Catch memory errors or other runtime issues
        err_msg = f"[GenerationError] {type(e).__name__}: {str(e)}"
        # Measure time until error
        end = time.perf_counter()
        latency_ms = (end - start) * 1000.0
        peak_mem_mb = None
        if device == "cuda":
            try:
                peak_mem_mb = torch.cuda.max_memory_allocated() / (1024 ** 2)
            except Exception:
                peak_mem_mb = None
        return err_msg, latency_ms, peak_mem_mb

    if device == "cuda":
        try:
            torch.cuda.synchronize()
        except Exception:
            pass

    end = time.perf_counter()
    latency_ms = (end - start) * 1000.0

    # Get only the generated completion (without repeating the prompt)
    input_len = inputs["input_ids"].shape[1]
    gen_ids = output[0][input_len:]
    if gen_ids.numel() == 0:
        raw_text = ""
    else:
        raw_text = tokenizer.decode(gen_ids, skip_special_tokens=True).strip()

    peak_mem_mb = None
    if device == "cuda":
        try:
            peak_mem_mb = torch.cuda.max_memory_allocated() / (1024 ** 2)
        except Exception:
            peak_mem_mb = None

    return raw_text, latency_ms, peak_mem_mb


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--model", default="Qwen/Qwen2.5-1.5B-Instruct")
    parser.add_argument("--device", default="auto", choices=["auto", "cuda", "cpu"])
    parser.add_argument("--max_new_tokens", type=int, default=32)
    parser.add_argument("--questions", default="questions.jsonl")
    parser.add_argument("--limit", type=int, default=2, help="Number of questions for quick testing (0 => all)")
    parser.add_argument("--out_csv", default="results.csv")
    args = parser.parse_args()

    device = pick_device(args.device)
    print(f"[Info] Using device: {device}")

    questions = read_questions_jsonl(args.questions)
    if args.limit > 0:
        questions = questions[: args.limit]

    if len(questions) == 0:
        print("[Error] No questions found in file:", args.questions)
        return

    # Load tokenizer + model (FP16 on GPU, FP32 on CPU)
    try:
        tokenizer = AutoTokenizer.from_pretrained(args.model, use_fast=True)
    except Exception as e:
        print("[Error] Failed to load tokenizer:", e)
        return

    if tokenizer.pad_token is None and tokenizer.eos_token is not None:
        tokenizer.pad_token = tokenizer.eos_token

    try:
        model = AutoModelForCausalLM.from_pretrained(
            args.model,
            torch_dtype=torch.float16 if device == "cuda" else torch.float32,
        ).to(device)
        model.eval()
    except Exception as e:
        print("[Error] Failed to load model:", e)
        return

    rows = []

    # Simple warmup to reduce initialization impact on measurement
    try:
        warm_prompt = build_prompts(questions[0]["question"])["simple"]
        _raw, _lat, _mem = generate_once(model, tokenizer, warm_prompt, device, max_new_tokens=8)
    except Exception:
        pass

    total = len(questions) * len(PROMPT_TYPES)
    counter = 0
    start_all = time.perf_counter()

    for q in questions:
        prompt_map = build_prompts(q["question"])
        for ptype in PROMPT_TYPES:
            counter += 1
            print(f"[{counter}/{total}] Q={q['id']} prompt={ptype}", flush=True)

            prompt = prompt_map[ptype]
            raw_text, latency, mem = generate_once(model, tokenizer, prompt, device, args.max_new_tokens)

            # Strict cleaning before saving
            clean_text = clean_output(raw_text)

            rows.append(
                {
                    "question_id": q["id"],
                    "prompt_type": ptype,
                    "precision_type": "fp16" if device == "cuda" else "fp32_cpu",
                    "latency_ms": latency,
                    "gpu_mem_mb": mem,
                    "raw_output_text": raw_text,
                    "output_text": clean_text,
                    # Rubric columns left empty for manual evaluation later
                    "correctness": None,
                    "completeness": None,
                    "clarity": None,
                    "final_score": None,
                }
            )

    elapsed = time.perf_counter() - start_all
    print(f"[Done] Finished {counter} generations in {elapsed:.1f}s")

    # Save results
    df = pd.DataFrame(rows)
    # final_score: leave empty for now; user will fill it manually
    df.to_csv(args.out_csv, index=False, encoding="utf-8-sig")
    print(f"[Info] results.csv written -> {args.out_csv}")


if __name__ == "__main__":
    main()
````

````python
%%writefile run_eval_4bit.py
# run_eval_4bit.py
# Evaluation of prompt sensitivity on 4-bit quantized model (bitsandbytes)

import argparse
import json
import time
from typing import Tuple, Optional, Dict, Any, List

import pandas as pd
import torch

from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from prompts import build_prompts, PROMPT_TYPES


def read_questions_jsonl(path: str) -> List[Dict[str, Any]]:
    """Reads questions from a JSONL file."""
    questions = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            questions.append(json.loads(line))
    return questions


def pick_device(device: str) -> str:
    """Picks the appropriate device (cuda or cpu)."""
    if device == "auto":
        return "cuda" if torch.cuda.is_available() else "cpu"
    return device


def clean_output(text: str) -> str:
    """Strict cleaning of model outputs (truncates dialogue/steps/examples, limits to two lines)."""
    if not text:
        return ""
    t = text.strip()

    cut_markers = [
        "Human:", "Assistant:", "USER:", "SYSTEM:",
        "المستخدم:", "المساعد:", "النظام:",
        "###", "---", "Human :", "Assistant :"
    ]
    for m in cut_markers:
        if m in t and not t.startswith(m):
            t = t.split(m, 1)[0].strip()

    example_markers = ["السؤال:", "سؤال المستخدم:", "سؤال:", "أمثلة:", "مثال:"]
    for m in example_markers:
        if m in t and not t.startswith(m):
            t = t.split(m, 1)[0].strip()

    step_markers = ["الخطوات", "خطوات", "فكرة الخطوات", "تفكير", "التفكير", "الخطوات الداخلية", "خطوة", "خطوات:"]
    for m in step_markers:
        if m in t:
            t = t.split(m, 1)[0].strip()

    prefixes_to_strip = ["إجابة:", "Answer:", "الإجابة:", "Result:", "النتيجة:"]
    for p in prefixes_to_strip:
        if t.startswith(p):
            t = t[len(p):].strip()

    t = t.replace("\r\n", "\n").replace("\r", "\n")
    while "\n\n" in t:
        t = t.replace("\n\n", "\n")
    t = t.strip()

    lines = [ln.strip() for ln in t.splitlines() if ln.strip()]
    if not lines:
        return ""
    t = "\n".join(lines[:2]).strip()

    if len(t) > 300:
        t = t[:297].rstrip() + "..."
    return t


def load_model_4bit(model_name: str):
    """Loads a 4-bit quantized model using bitsandbytes."""
    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_use_double_quant=True,
        bnb_4bit_compute_dtype=torch.float16,
    )

    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        quantization_config=bnb_config,
        device_map="auto",
    )
    model.eval()
    return model


@torch.inference_mode()
def generate_once(model, tokenizer, prompt: str, device: str, max_new_tokens: int) -> Tuple[str, float, Optional[float]]:
    """Generates text from the model and measures latency and GPU memory."""
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)

    if device == "cuda":
        inputs = {k: v.to("cuda") for k, v in inputs.items()}
        try:
            torch.cuda.reset_peak_memory_stats()
            torch.cuda.synchronize()
        except Exception:
            pass
    else:
        inputs = {k: v.to("cpu") for k, v in inputs.items()}

    start = time.perf_counter()

    try:
        out = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            temperature=0.0,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )
    except RuntimeError as e:
        end = time.perf_counter()
        latency_ms = (end - start) * 1000.0
        mem = None
        if device == "cuda":
            try:
                mem = torch.cuda.max_memory_allocated() / (1024 ** 2)
            except Exception:
                mem = None
        return f"[GenerationError] {type(e).__name__}: {str(e)}", latency_ms, mem

    if device == "cuda":
        try:
            torch.cuda.synchronize()
        except Exception:
            pass

    end = time.perf_counter()
    latency_ms = (end - start) * 1000.0

    input_len = inputs["input_ids"].shape[1]
    gen_ids = out[0][input_len:]
    raw_text = tokenizer.decode(gen_ids, skip_special_tokens=True).strip() if gen_ids.numel() else ""

    mem = None
    if device == "cuda":
        try:
            mem = torch.cuda.max_memory_allocated() / (1024 ** 2)
        except Exception:
            mem = None

    return raw_text, latency_ms, mem


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--model", default="Qwen/Qwen2.5-1.5B-Instruct")
    parser.add_argument("--device", default="auto", choices=["auto", "cuda", "cpu"])
    parser.add_argument("--max_new_tokens", type=int, default=46)
    parser.add_argument("--questions", default="questions.jsonl")
    parser.add_argument("--limit", type=int, default=30, help="0 => all questions")
    parser.add_argument("--out_csv", default="results_4bit.csv")
    args = parser.parse_args()

    device = pick_device(args.device)
    if device != "cuda":
        print("[Warning] 4-bit quantization with bitsandbytes requires a GPU (cuda). It may not work on CPU.")
    print(f"[Info] Using device: {device}")

    questions = read_questions_jsonl(args.questions)
    if args.limit > 0:
        questions = questions[:args.limit]
    if not questions:
        print("[Error] No questions found.")
        return

    tokenizer = AutoTokenizer.from_pretrained(args.model, use_fast=True)
    if tokenizer.pad_token is None and tokenizer.eos_token is not None:
        tokenizer.pad_token = tokenizer.eos_token

    if device != "cuda":
        print("[Error] Please run on GPU: --device cuda or --device auto with GPU enabled.")
        return

    print("[Info] Loading 4-bit model...")
    model = load_model_4bit(args.model)

    rows = []

    # Warmup
    try:
        warm_prompt = build_prompts(questions[0]["question"])["simple"]
        _ = generate_once(model, tokenizer, warm_prompt, device, max_new_tokens=8)
    except Exception:
        pass

    total = len(questions) * len(PROMPT_TYPES)
    c = 0

    for q in questions:
        prompt_map = build_prompts(q["question"])
        for ptype in PROMPT_TYPES:
            c += 1
            print(f"[{c}/{total}] Q={q['id']} prompt={ptype}", flush=True)

            prompt = prompt_map[ptype]
            raw_text, latency, mem = generate_once(model, tokenizer, prompt, device, args.max_new_tokens)
            clean_text = clean_output(raw_text)

            rows.append({
                "question_id": q["id"],
                "prompt_type": ptype,
                "precision_type": "4bit",
                "latency_ms": latency,
                "gpu_mem_mb": mem,
                "raw_output_text": raw_text,
                "output_text": clean_text,
                "correctness": None,
                "completeness": None,
                "clarity": None,
                "final_score": None,
            })

    pd.DataFrame(rows).to_csv(args.out_csv, index=False, encoding="utf-8-sig")
    print(f"[Info] 4-bit results written -> {args.out_csv}")


if __name__ == "__main__":
    main()
````

````python
%%writefile plot_results.py
# -*- coding: utf-8 -*-
"""
End-to-end evaluation + analysis for fp16 vs 4bit using questions.jsonl

Inputs:
- questions.jsonl
- results.csv          (fp16)
- results_4bit.csv     (4bit)

Outputs:
- results_scored_fp16_all30_verified.csv
- results_scored_4bit_all30_verified.csv
- summary_fp16_vs_4bit_all30.csv
- compare_box_final_score.png
- compare_latency.png
- compare_gpu_mem.png
- compare_scatter_latency_score.png
"""

import json
import re
import math
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display


# =========================
# Paths (EDIT IF NEEDED)
# =========================
QUESTIONS_PATH = Path("questions.jsonl")
FP16_PATH = Path("results.csv")
Q4BIT_PATH = Path("results_4bit.csv")

OUT_FP16 = Path("results_scored_fp16_all30_verified.csv")
OUT_4BIT = Path("results_scored_4bit_all30_verified.csv")
OUT_SUMMARY = Path("summary_fp16_vs_4bit_all30.csv")

PLOT_BOX = Path("compare_box_final_score.png")
PLOT_LAT = Path("compare_latency.png")
PLOT_MEM = Path("compare_gpu_mem.png")
PLOT_SCAT = Path("compare_scatter_latency_score.png")


# =========================
# Utils
# =========================
def load_questions(path: Path) -> dict:
    """Loads questions from a JSONL file into a dictionary."""
    questions = {}
    with path.open("r", encoding="utf-8") as f:
        for line in f:
            obj = json.loads(line)
            questions[obj["id"]] = obj["question"]
    return questions


def norm(s: str) -> str:
    """Normalizes Arabic/Latin text for robust keyword matching (lowercase, remove punctuation, reduce whitespace)."""
    if not isinstance(s, str):
        return ""
    s = s.lower()
    # keep arabic/latin/digits/underscore; everything else => space
    s = re.sub(r"[^\w\u0600-\u06FF]+", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s


def has_leak(s: str) -> bool:
    """Detects prompt leakage such as 'Human:' or 'Assistant:' in the generated text."""
    if not isinstance(s, str):
        return False
    return bool(re.search(r"\b(human|assistant)\s*:", s, flags=re.I))


def contains_any(ans_norm: str, patterns) -> bool:
    """Checks if the normalized answer contains any of the given patterns (strings or regex)."""
    for p in patterns:
        if isinstance(p, str):
            # if looks like regex word-boundary pattern, treat as regex
            if "\b" in p or "[" in p or "(" in p:
                if re.search(p, ans_norm):
                    return True
            else:
                if norm(p) in ans_norm:
                    return True
        else:
            if re.search(p, ans_norm):
                return True
    return False


# =========================
# Rubric / Answer Key (30 Qs)
# =========================
# Defines the expected answers and scoring criteria for each question.
RUBRIC = {
    "Q1":  {"type": "single",     "accept": ["القاهرة"]},
    "Q2":  {"type": "single",     "accept": [r"\bh2o\b"]},
    "Q3":  {"type": "single",     "accept": [r"\b8\b", "ثماني", "ثمانية"]},
    "Q4":  {"type": "single",     "accept": ["البناء الضوئي", "التمثيل الضوئي", "التركيب الضوئي"]},
    "Q5":  {"type": "single",     "accept": [r"\b3\b", "ثلاث"]},
    "Q6":  {"type": "single",     "accept": ["المحيط الهادئ", "الهادئ"]},
    "Q7":  {"type": "single",     "accept": ["النيتروجين", "النتروجين", "nitrogen"]},
    "Q8":  {"type": "single",     "accept": ["القلب"]},

    # Definitions / explanations: must-hit keyword groups
    "Q9":  {"type": "definition", "must": [["الجاذبية", "gravity"],
                                          ["انحناء", "تقوس", "curvature"],
                                          ["الزمكان", "الفضاء", "الزمان", "spacetime"]],
            "optional": ["اينشتاين", "النسبية"]},

    "Q10": {"type": "numeric",    "accept": [r"\b18000000\b", r"\b18\s*000\s*000\b", "18 مليون", "ثمانية عشر مليون"],
            "unit": ["كيلومتر", "كم", "km"]},

    "Q11": {"type": "single",     "accept": [r"\b8\b", "ثمانية"]},
    "Q12": {"type": "single",     "accept": ["الذرة", "atom"]},

    "Q13": {"type": "compare",    "must": [["جدار"],
                                          ["كلوروبلاست", "بلاستيدات", "chloroplast"],
                                          ["فجوة", "vacuole"]]} ,

    "Q14": {"type": "definition", "must": [["طاقة", "energy"],
                                          ["atp", "أ تي بي"]],
            "optional": ["تنفس", "خلوي"]},

    "Q15": {"type": "single",     "accept": [r"\b14\b", "أربعة عشر", "اربعة عشر"]},

    "Q16": {"type": "definition", "must": [["القوة", "force", "f"],
                                          ["الكتلة", "mass", "m"],
                                          ["التسارع", "acceleration", "a"]],
            "optional": [r"\bf\s*=\s*m\s*a\b"]},

    "Q17": {"type": "definition", "must": [["معلومات", "information"],
                                          ["وراث", "جين", "gene"],
                                          ["خلايا", "نواة", "cell"]],
            "optional": ["dna", "دنا", "الحمض النووي"]},

    # Note: Q18 has multiple common answers online; we accept common candidates
    "Q18": {"type": "single",     "accept": ["الدلفين", "دولفين", "القرش", "سمك القرش", "dolphin", "shark"]},

    "Q19": {"type": "single",     "accept": [r"\b72\b", "اثنان وسبعون", "اثنين وسبعين"]},
    "Q20": {"type": "single",     "accept": ["المريخ", "mars"]},
    "Q21": {"type": "single",     "accept": [r"\bnacl\b", "كلوريد الصوديوم"]},
    "Q22": {"type": "single",     "accept": ["النعامة", "البطريق", "الكيوي", "emu", "ostrich", "penguin", "kiwi"]},
    "Q23": {"type": "single",     "accept": [r"\b16\b", "ستة عشر"], "unit": ["سم", "cm", "سنتيمتر"]},
    "Q24": {"type": "single",     "accept": ["الفهد", "cheetah", "الشاهين", "peregrine"]},
    "Q25": {"type": "single",     "accept": ["الهيدروجين", "hydrogen"]},
    "Q26": {"type": "single",     "accept": ["البنكرياس", "pancreas"]},
    "Q27": {"type": "single",     "accept": [r"\b5\b", "خمسة"], "optional": ["فيثاغورس"]},
    "Q28": {"type": "definition", "must": [["حركة", "تحرك"],
                                          ["صفائح", "تكتونية", "tectonic"],
                                          ["فالق", "صدع", "fault"]],
            "optional": ["اجهاد", "إجهاد", "طاقة"]},
    "Q29": {"type": "single",     "accept": ["الفلسبار", "feldspar"]},
    "Q30": {"type": "single",     "accept": ["التخمر", "التخمير", "fermentation"]},
}


# =========================
# Scoring Functions
# =========================
def score_clarity(txt: str) -> float:
    """Scores the clarity of the output text based on length and presence of prompt leakage indicators."""
    if not isinstance(txt, str) or not txt.strip():
        return 0.0
    score = 5.0
    if has_leak(txt):
        score -= 2.0
    # overly long answers penalized a bit
    if len(txt.split()) > 80:
        score -= 1.0
    # "internal steps" style penalty
    if re.search(r"الخطوات الداخلية|تفكير داخلي", txt):
        score -= 1.0
    return float(max(0.0, score))


def score_one_row(qid: str, output_text: str) -> tuple[float, float, float]:
    """
    Scores a single model response based on correctness, completeness, and clarity.
    Returns: (correctness, completeness, clarity) in [0..5]
    """
    spec = RUBRIC.get(qid)
    clarity = score_clarity(output_text)

    if spec is None:
        # Unknown question rubric: can't verify
        return 0.0, 0.0, clarity

    a = norm(output_text)
    leak = has_leak(output_text)

    correctness = 0.0
    completeness = 0.0

    t = spec["type"]

    if t in ("single", "numeric"):
        ok = contains_any(a, spec["accept"])
        if ok and spec.get("unit"):
            ok = ok and contains_any(a, spec["unit"])
        correctness = 5.0 if ok else 0.0
        completeness = 5.0 if ok else 1.0

    elif t == "definition":
        must_groups = spec.get("must", [])
        hits = sum(1 for g in must_groups if contains_any(a, g))
        # correctness: all must groups -> 5, almost all -> 3, else 0
        if hits >= len(must_groups):
            correctness = 5.0
        elif hits >= max(1, len(must_groups) - 1):
            correctness = 3.0
        else:
            correctness = 0.0
        completeness = 5.0 * (hits / len(must_groups)) if must_groups else (5.0 if correctness > 0 else 1.0)
        if correctness == 0.0:
            completeness = 1.0

    elif t == "compare":
        must_groups = spec.get("must", [])
        hits = sum(1 for g in must_groups if contains_any(a, g))
        if hits >= 2:
            correctness = 5.0
        elif hits == 1:
            correctness = 3.0
        else:
            correctness = 0.0
        completeness = 5.0 * (hits / len(must_groups)) if must_groups else 1.0
        if correctness == 0.0:
            completeness = 1.0

    # optional penalty: prompt leakage tends to mean broken output
    if leak and correctness > 0:
        correctness = max(0.0, correctness - 1.0)

    return float(np.clip(correctness, 0, 5)), float(np.clip(completeness, 0, 5)), float(np.clip(clarity, 0, 5))


def score_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    """Applies scoring to each row of the DataFrame and calculates final_score."""
    df = df.copy()
    cors, comps, clars = [], [], []
    for qid, txt in zip(df["question_id"].astype(str), df["output_text"].astype(str)):
        c, co, cl = score_one_row(qid, txt)
        cors.append(c); comps.append(co); clars.append(cl)

    df["correctness"] = cors
    df["completeness"] = comps
    df["clarity"] = clars
    df["final_score"] = df["correctness"] * 0.5 + df["completeness"] * 0.3 + df["clarity"] * 0.2
    return df


# =========================
# Main Execution and Plotting
# =========================
def main():
    """Loads results, scores them, and generates comparison plots for FP16 vs 4-bit models."""
    # Load questions
    questions = load_questions(QUESTIONS_PATH)

    # Load FP16 and 4-bit results
    fp16 = pd.read_csv(FP16_PATH)
    fp16 = fp16[fp16["precision_type"].astype(str).str.lower() == "fp16"].copy()
    fp16["question_text"] = fp16["question_id"].map(questions)

    q4 = pd.read_csv(Q4BIT_PATH)
    # keep only quantized rows
    q4 = q4[q4["precision_type"].astype(str).str.lower().str.contains("4")].copy()
    q4["question_text"] = q4["question_id"].map(questions)

    # Score the dataframes
    fp16_scored = score_dataframe(fp16)
    q4_scored = score_dataframe(q4)

    # Save scored outputs
    fp16_scored.to_csv(OUT_FP16, index=False, encoding="utf-8-sig")
    q4_scored.to_csv(OUT_4BIT, index=False, encoding="utf-8-sig")

    # Combine for summary and plotting
    fp16_scored["precision_type"] = "fp16"
    q4_scored["precision_type"] = "4bit"
    all_df = pd.concat([fp16_scored, q4_scored], ignore_index=True)

    # Generate summary statistics
    summary = (
        all_df.groupby(["prompt_type", "precision_type"])
              .agg(
                  final_score_mean=("final_score", "mean"),
                  final_score_std=("final_score", "std"),
                  correctness_mean=("correctness", "mean"),
                  latency_ms_mean=("latency_ms", "mean"),
                  gpu_mem_mb_mean=("gpu_mem_mb", "mean"),
                  n=("final_score", "size"),
              )
              .reset_index()
    )
    summary["balance_metric"] = summary["final_score_mean"] / np.log(summary["latency_ms_mean"])
    summary.to_csv(OUT_SUMMARY, index=False, encoding="utf-8-sig")

    # ===== PLOTS =====
    prompt_types = sorted(all_df["prompt_type"].unique())
    precision_types = ["fp16", "4bit"]

    # Boxplot: final_score by prompt_type split by precision
    fig, ax = plt.subplots(figsize=(12, 5))
    width = 0.35
    positions, data = [], []
    for i, p in enumerate(prompt_types):
        for j, prec in enumerate(precision_types):
            subset = all_df[(all_df["prompt_type"] == p) & (all_df["precision_type"] == prec)]["final_score"]
            positions.append(i + (j - 0.5) * width)
            data.append(subset.values)
    ax.boxplot(data, positions=positions, widths=width)
    ax.set_xticks(range(len(prompt_types)))
    ax.set_xticklabels(prompt_types, rotation=25, ha="right")
    ax.set_ylabel("final_score")
    ax.set_title("Final Score Distribution (fp16 vs 4bit)")
    ax.grid(True, axis="y", alpha=0.3)
    fig.tight_layout()
    fig.savefig(PLOT_BOX, dpi=200)
    plt.close(fig)

    # Bar: latency
    lat = summary.pivot(index="prompt_type", columns="precision_type", values="latency_ms_mean").loc[prompt_types]
    fig, ax = plt.subplots(figsize=(12, 5))
    lat.plot(kind="bar", ax=ax)
    ax.set_ylabel("mean latency_ms")
    ax.set_title("Latency Comparison fp16 vs 4bit")
    ax.grid(True, axis="y", alpha=0.3)
    fig.tight_layout()
    fig.savefig(PLOT_LAT, dpi=200)
    plt.close(fig)

    # Bar: GPU mem
    mem = summary.pivot(index="prompt_type", columns="precision_type", values="gpu_mem_mb_mean").loc[prompt_types]
    fig, ax = plt.subplots(figsize=(12, 5))
    mem.plot(kind="bar", ax=ax)
    ax.set_ylabel("mean gpu_mem_mb")
    ax.set_title("GPU Memory Comparison fp16 vs 4bit")
    ax.grid(True, axis="y", alpha=0.3)
    fig.tight_layout()
    fig.savefig(PLOT_MEM, dpi=200)
    plt.close(fig)

    # Scatter: latency vs score (group means)
    fig, ax = plt.subplots(figsize=(8, 6))
    for prec, marker in zip(["fp16", "4bit"], ["o", "s"]):
        sub = summary[summary["precision_type"] == prec]
        ax.scatter(sub["latency_ms_mean"], sub["final_score_mean"], marker=marker, label=prec)
        for _, r in sub.iterrows():
            ax.text(r["latency_ms_mean"], r["final_score_mean"], r["prompt_type"], fontsize=8)
    ax.set_xlabel("latency_ms_mean")
    ax.set_ylabel("final_score_mean")
    ax.set_title("Latency vs Score (Group Means)")
    ax.grid(True, alpha=0.3)
    ax.legend()
    fig.tight_layout()
    fig.savefig(PLOT_SCAT, dpi=200)
    plt.close(fig)

    print("DONE ✅")
    print(f"- {OUT_FP16}")
    print(f"- {OUT_4BIT}")
    print(f"- {OUT_SUMMARY}")
    print(f"- {PLOT_BOX}")
    print(f"- {PLOT_LAT}")
    print(f"- {PLOT_MEM}")
    print(f"- {PLOT_SCAT}")

    # Display the plots
    print("\n--- Generated Plots ---")
    for plot_path in [PLOT_BOX, PLOT_LAT, PLOT_MEM, PLOT_SCAT]:
        if plot_path.exists():
            print(f"Displaying: {plot_path}")
            display(Image(filename=plot_path))
        else:
            print(f"Plot not found: {plot_path}")


if __name__ == "__main__":
    main()
````

````python
# -*- coding: utf-8 -*-
"""
End-to-end evaluation + analysis for fp16 vs 4bit using questions.jsonl

Inputs:
- questions.jsonl
- results.csv          (fp16)
- results_4bit.csv     (4bit)

Outputs:
- results_scored_fp16_all30_verified.csv
- results_scored_4bit_all30_verified.csv
- summary_fp16_vs_4bit_all30.csv
- compare_box_final_score.png
- compare_latency.png
- compare_gpu_mem.png
- compare_scatter_latency_score.png
"""

import json
import re
import math
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display


# =========================
# Paths (EDIT IF NEEDED)
# =========================
QUESTIONS_PATH = Path("questions.jsonl")
FP16_PATH = Path("results.csv")
Q4BIT_PATH = Path("results_4bit.csv")

OUT_FP16 = Path("results_scored_fp16_all30_verified.csv")
OUT_4BIT = Path("results_scored_4bit_all30_verified.csv")
OUT_SUMMARY = Path("summary_fp16_vs_4bit_all30.csv")

PLOT_BOX = Path("compare_box_final_score.png")
PLOT_LAT = Path("compare_latency.png")
PLOT_MEM = Path("compare_gpu_mem.png")
PLOT_SCAT = Path("compare_scatter_latency_score.png")


# =========================
# Utils
# =========================
def load_questions(path: Path) -> dict:
    questions = {}
    with path.open("r", encoding="utf-8") as f:
        for line in f:
            obj = json.loads(line)
            questions[obj["id"]] = obj["question"]
    return questions


def norm(s: str) -> str:
    """Normalize Arabic/Latin text for robust matching."""
    if not isinstance(s, str):
        return ""
    s = s.lower()
    # keep arabic/latin/digits/underscore; everything else => space
    s = re.sub(r"[^\w\u0600-\u06FF]+", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s


def has_leak(s: str) -> bool:
    """Detect prompt leakage such as Human:/Assistant:"""
    if not isinstance(s, str):
        return False
    return bool(re.search(r"\b(human|assistant)\s*:", s, flags=re.I))


def contains_any(ans_norm: str, patterns) -> bool:
    """patterns: list of strings or regex patterns (as raw strings with \b etc)."""
    for p in patterns:
        if isinstance(p, str):
            # if looks like regex word-boundary pattern, treat as regex
            if "\b" in p or "[" in p or "(" in p:
                if re.search(p, ans_norm):
                    return True
            else:
                if norm(p) in ans_norm:
                    return True
        else:
            if re.search(p, ans_norm):
                return True
    return False


# =========================
# Rubric / Answer Key (30 Qs)
# =========================
RUBRIC = {
    "Q1":  {"type": "single",     "accept": ["القاهرة"]},
    "Q2":  {"type": "single",     "accept": [r"\bh2o\b"]},
    "Q3":  {"type": "single",     "accept": [r"\b8\b", "ثماني", "ثمانية"]},
    "Q4":  {"type": "single",     "accept": ["البناء الضوئي", "التمثيل الضوئي", "التركيب الضوئي"]},
    "Q5":  {"type": "single",     "accept": [r"\b3\b", "ثلاث"]},
    "Q6":  {"type": "single",     "accept": ["المحيط الهادئ", "الهادئ"]},
    "Q7":  {"type": "single",     "accept": ["النيتروجين", "النتروجين", "nitrogen"]},
    "Q8":  {"type": "single",     "accept": ["القلب"]},

    # Definitions / explanations: must-hit keyword groups
    "Q9":  {"type": "definition", "must": [["الجاذبية", "gravity"],
                                          ["انحناء", "تقوس", "curvature"],
                                          ["الزمكان", "الفضاء", "الزمان", "spacetime"]],
            "optional": ["اينشتاين", "النسبية"]},

    "Q10": {"type": "numeric",    "accept": [r"\b18000000\b", r"\b18\s*000\s*000\b", "18 مليون", "ثمانية عشر مليون"],
            "unit": ["كيلومتر", "كم", "km"]},

    "Q11": {"type": "single",     "accept": [r"\b8\b", "ثمانية"]},
    "Q12": {"type": "single",     "accept": ["الذرة", "atom"]},

    "Q13": {"type": "compare",    "must": [["جدار"],
                                          ["كلوروبلاست", "بلاستيدات", "chloroplast"],
                                          ["فجوة", "vacuole"]]},

    "Q14": {"type": "definition", "must": [["طاقة", "energy"],
                                          ["atp", "أ تي بي"]],
            "optional": ["تنفس", "خلوي"]},

    "Q15": {"type": "single",     "accept": [r"\b14\b", "أربعة عشر", "اربعة عشر"]},

    "Q16": {"type": "definition", "must": [["القوة", "force", "f"],
                                          ["الكتلة", "mass", "m"],
                                          ["التسارع", "acceleration", "a"]],
            "optional": [r"\bf\s*=\s*m\s*a\b"]},

    "Q17": {"type": "definition", "must": [["معلومات", "information"],
                                          ["وراث", "جين", "gene"],
                                          ["خلايا", "نواة", "cell"]],
            "optional": ["dna", "دنا", "الحمض النووي"]},

    # Note: Q18 has multiple common answers online; we accept common candidates
    "Q18": {"type": "single",     "accept": ["الدلفين", "دولفين", "القرش", "سمك القرش", "dolphin", "shark"]},

    "Q19": {"type": "single",     "accept": [r"\b72\b", "اثنان وسبعون", "اثنين وسبعين"]},
    "Q20": {"type": "single",     "accept": ["المريخ", "mars"]},
    "Q21": {"type": "single",     "accept": [r"\bnacl\b", "كلوريد الصوديوم"]},
    "Q22": {"type": "single",     "accept": ["النعامة", "البطريق", "الكيوي", "emu", "ostrich", "penguin", "kiwi"]},
    "Q23": {"type": "single",     "accept": [r"\b16\b", "ستة عشر"], "unit": ["سم", "cm", "سنتيمتر"]},
    "Q24": {"type": "single",     "accept": ["الفهد", "cheetah", "الشاهين", "peregrine"]},
    "Q25": {"type": "single",     "accept": ["الهيدروجين", "hydrogen"]},
    "Q26": {"type": "single",     "accept": ["البنكرياس", "pancreas"]},
    "Q27": {"type": "single",     "accept": [r"\b5\b", "خمسة"], "optional": ["فيثاغورس"]},
    "Q28": {"type": "definition", "must": [["حركة", "تحرك"],
                                          ["صفائح", "تكتونية", "tectonic"],
                                          ["فالق", "صدع", "fault"]],
            "optional": ["اجهاد", "إجهاد", "طاقة"]},
    "Q29": {"type": "single",     "accept": ["الفلسبار", "feldspar"]},
    "Q30": {"type": "single",     "accept": ["التخمر", "التخمير", "fermentation"]},
}


# =========================
# Scoring
# =========================
def score_clarity(txt: str) -> float:
    if not isinstance(txt, str) or not txt.strip():
        return 0.0
    score = 5.0
    if has_leak(txt):
        score -= 2.0
    # overly long answers penalized a bit
    if len(txt.split()) > 80:
        score -= 1.0
    # "internal steps" style penalty
    if re.search(r"الخطوات الداخلية|تفكير داخلي", txt):
        score -= 1.0
    return float(max(0.0, score))


def score_one_row(qid: str, output_text: str) -> tuple[float, float, float]:
    """
    Returns: correctness, completeness, clarity in [0..5]
    Rule:
    - For single/numeric: correctness 5 if match else 0, completeness 5 if correct else 1
    - For definition/compare: graded by keyword group coverage
    """
    spec = RUBRIC.get(qid)
    clarity = score_clarity(output_text)

    if spec is None:
        # Unknown question rubric: can't verify
        return 0.0, 0.0, clarity

    a = norm(output_text)
    leak = has_leak(output_text)

    correctness = 0.0
    completeness = 0.0

    t = spec["type"]

    if t in ("single", "numeric"):
        ok = contains_any(a, spec["accept"])
        if ok and spec.get("unit"):
            ok = ok and contains_any(a, spec["unit"])
        correctness = 5.0 if ok else 0.0
        completeness = 5.0 if ok else 1.0

    elif t == "definition":
        must_groups = spec.get("must", [])
        hits = sum(1 for g in must_groups if contains_any(a, g))
        # correctness: all must groups -> 5, almost all -> 3, else 0
        if hits >= len(must_groups):
            correctness = 5.0
        elif hits >= max(1, len(must_groups) - 1):
            correctness = 3.0
        else:
            correctness = 0.0
        completeness = 5.0 * (hits / len(must_groups)) if must_groups else (5.0 if correctness > 0 else 1.0)
        if correctness == 0.0:
            completeness = 1.0

    elif t == "compare":
        must_groups = spec.get("must", [])
        hits = sum(1 for g in must_groups if contains_any(a, g))
        if hits >= 2:
            correctness = 5.0
        elif hits == 1:
            correctness = 3.0
        else:
            correctness = 0.0
        completeness = 5.0 * (hits / len(must_groups)) if must_groups else 1.0
        if correctness == 0.0:
            completeness = 1.0

    # optional penalty: prompt leakage tends to mean broken output
    if leak and correctness > 0:
        correctness = max(0.0, correctness - 1.0)

    return float(np.clip(correctness, 0, 5)), float(np.clip(completeness, 0, 5)), float(np.clip(clarity, 0, 5))


def score_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    cors, comps, clars = [], [], []
    for qid, txt in zip(df["question_id"].astype(str), df["output_text"].astype(str)):
        c, co, cl = score_one_row(qid, txt)
        cors.append(c); comps.append(co); clars.append(cl)

    df["correctness"] = cors
    df["completeness"] = comps
    df["clarity"] = clars
    df["final_score"] = df["correctness"] * 0.5 + df["completeness"] * 0.3 + df["clarity"] * 0.2
    return df


# =========================
# Main
# =========================
def main():
    # Load questions
    questions = load_questions(QUESTIONS_PATH)

    # Load fp16 + 4bit
    fp16 = pd.read_csv(FP16_PATH)
    fp16 = fp16[fp16["precision_type"].astype(str).str.lower() == "fp16"].copy()
    fp16["question_text"] = fp16["question_id"].map(questions)

    q4 = pd.read_csv(Q4BIT_PATH)
    # keep only quantized rows
    q4 = q4[q4["precision_type"].astype(str).str.lower().str.contains("4")].copy()
    q4["question_text"] = q4["question_id"].map(questions)

    # Score
    fp16_scored = score_dataframe(fp16)
    q4_scored = score_dataframe(q4)

    # Save scored outputs
    fp16_scored.to_csv(OUT_FP16, index=False, encoding="utf-8-sig")
    q4_scored.to_csv(OUT_4BIT, index=False, encoding="utf-8-sig")

    # Combine summary
    fp16_scored["precision_type"] = "fp16"
    q4_scored["precision_type"] = "4bit"
    all_df = pd.concat([fp16_scored, q4_scored], ignore_index=True)

    summary = (
        all_df.groupby(["prompt_type", "precision_type"])
              .agg(
                  final_score_mean=("final_score", "mean"),
                  final_score_std=("final_score", "std"),
                  correctness_mean=("correctness", "mean"),
                  latency_ms_mean=("latency_ms", "mean"),
                  gpu_mem_mb_mean=("gpu_mem_mb", "mean"),
                  n=("final_score", "size"),
              )
              .reset_index()
    )
    summary["balance_metric"] = summary["final_score_mean"] / np.log(summary["latency_ms_mean"])
    summary.to_csv(OUT_SUMMARY, index=False, encoding="utf-8-sig")

    # ===== PLOTS =====
    prompt_types = sorted(all_df["prompt_type"].unique())
    precision_types = ["fp16", "4bit"]

    # Boxplot: final_score by prompt_type split by precision
    fig, ax = plt.subplots(figsize=(12, 5))
    width = 0.35
    positions, data = [], []
    for i, p in enumerate(prompt_types):
        for j, prec in enumerate(precision_types):
            subset = all_df[(all_df["prompt_type"] == p) & (all_df["precision_type"] == prec)]["final_score"]
            positions.append(i + (j - 0.5) * width)
            data.append(subset.values)
    ax.boxplot(data, positions=positions, widths=width)
    ax.set_xticks(range(len(prompt_types)))
    ax.set_xticklabels(prompt_types, rotation=25, ha="right")
    ax.set_ylabel("final_score")
    ax.set_title("Final Score Distribution (fp16 vs 4bit)")
    ax.grid(True, axis="y", alpha=0.3)
    fig.tight_layout()
    fig.savefig(PLOT_BOX, dpi=200)
    plt.close(fig)

    # Bar: latency
    lat = summary.pivot(index="prompt_type", columns="precision_type", values="latency_ms_mean").loc[prompt_types]
    fig, ax = plt.subplots(figsize=(12, 5))
    lat.plot(kind="bar", ax=ax)
    ax.set_ylabel("mean latency_ms")
    ax.set_title("Latency Comparison fp16 vs 4bit")
    ax.grid(True, axis="y", alpha=0.3)
    fig.tight_layout()
    fig.savefig(PLOT_LAT, dpi=200)
    plt.close(fig)

    # Bar: GPU mem
    mem = summary.pivot(index="prompt_type", columns="precision_type", values="gpu_mem_mb_mean").loc[prompt_types]
    fig, ax = plt.subplots(figsize=(12, 5))
    mem.plot(kind="bar", ax=ax)
    ax.set_ylabel("mean gpu_mem_mb")
    ax.set_title("GPU Memory Comparison fp16 vs 4bit")
    ax.grid(True, axis="y", alpha=0.3)
    fig.tight_layout()
    fig.savefig(PLOT_MEM, dpi=200)
    plt.close(fig)

    # Scatter: latency vs score (group means)
    fig, ax = plt.subplots(figsize=(8, 6))
    for prec, marker in zip(["fp16", "4bit"], ["o", "s"]):
        sub = summary[summary["precision_type"] == prec]
        ax.scatter(sub["latency_ms_mean"], sub["final_score_mean"], marker=marker, label=prec)
        for _, r in sub.iterrows():
            ax.text(r["latency_ms_mean"], r["final_score_mean"], r["prompt_type"], fontsize=8)
    ax.set_xlabel("latency_ms_mean")
    ax.set_ylabel("final_score_mean")
    ax.set_title("Latency vs Score (Group Means)")
    ax.grid(True, alpha=0.3)
    ax.legend()
    fig.tight_layout()
    fig.savefig(PLOT_SCAT, dpi=200)
    plt.close(fig)

    print("DONE ✅")
    print(f"- {OUT_FP16}")
    print(f"- {OUT_4BIT}")
    print(f"- {OUT_SUMMARY}")
    print(f"- {PLOT_BOX}")
    print(f"- {PLOT_LAT}")
    print(f"- {PLOT_MEM}")
    print(f"- {PLOT_SCAT}")

    # Display the plots
    print("\n--- Generated Plots ---")
    for plot_path in [PLOT_BOX, PLOT_LAT, PLOT_MEM, PLOT_SCAT]:
        if plot_path.exists():
            print(f"Displaying: {plot_path}")
            display(Image(filename=plot_path))
        else:
            print(f"Plot not found: {plot_path}")


if __name__ == "__main__":
    main()
````
